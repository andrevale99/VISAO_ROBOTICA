```python
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import os

PATH_IMG = "../IMAGENS/"

PATH = os.getcwd()

print(os.getcwd())
print(os.listdir(PATH_IMG))
print(cv.__version__)

img_original = cv.imread(PATH_IMG + os.listdir(PATH_IMG)[1])
img_gray = cv.cvtColor(img_original, cv.COLOR_BGR2GRAY)
```

# Imagem Canny

```python
img_gauss = cv.GaussianBlur(img_gray, (5, 5), 0)
img_canny = cv.Canny(img_gauss, 60, 90)

plt.imshow(cv.cvtColor(img_canny, cv.COLOR_BGR2RGB))
plt.show()
```
